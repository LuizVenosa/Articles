```python
votacoes_votos_dep['voto_numerico'] = votacoes_votos_dep['voto'].map({'Sim': 1, 'Não': 0})

# Pivot the data
votes_pivot = votacoes_votos_dep.pivot_table(index='deputado_nome', columns='idVotacao', values='voto_numerico')
```

```python
votes_pivot.isna().sum().sort_values(ascending=False)
```

```python
votes_pivot = votes_pivot.dropna(thresh=150, axis=1)
votes_pivot = votes_pivot.dropna(thresh=200, axis=0)
```

```python
votes_pivot.fillna(0.5, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
similarity_matrix = cosine_similarity(votes_pivot)
similarity_df = pd.DataFrame(similarity_matrix, index=votes_pivot.index, columns=votes_pivot.index)
similarity_df.to_csv('similarity_matrix.csv')
```

```python
import plotly.graph_objects as go
import networkx as nx

names = similarity_df.columns
# Create the graph as before
G = nx.Graph()
for i, name in enumerate(names):
    G.add_node(name)
counter = { name:[] for name in names}
# Add edges with a threshold
threshold = 0.70
for i in range(len(similarity_matrix)):
    for j in range(i + 1, len(similarity_matrix)):
        if similarity_matrix[i][j] > threshold:
            # G.add_edge(names[i], names[j], weight=similarity_matrix[i][j])
            counter[names[i]].append((names[j], similarity_matrix[i][j]))
for source, target in counter.items():
    selected_targets = sorted(target, key=lambda x: x[1], reverse=True)[:10]
    for target, weight in selected_targets:
        G.add_edge(source, target, weight=weight)
            

# Position nodes using NetworkX's layout
pos = nx.spring_layout(G, k=0.2,  iterations=50)
```

```python
print(len(G.edges))
```

```python

node_sizes = nx.degree_centrality(G)
# Create Edges
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

# Add edges as a scatter plot
edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=0.5, color='#888'), hoverinfo='none', mode='lines')
# Create Nodes
node_x = []
node_y = []
for node in G.nodes():
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)


# Add nodes as a scatter plot
node_trace = go.Scatter(x=node_x, y=node_y, mode='markers+text', hoverinfo='text', marker=dict(showscale=True, colorscale='YlGnBu', size=[(node_sizes[node]+1)*10 for node in G.nodes()], color=[], line_width=2))

# Add text to the nodes
node_trace.text = list(G.nodes())

# Create a figure
fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(showlegend=False, hovermode='closest', margin=dict(b=0,l=0,r=0,t=0), xaxis=dict(showgrid=False, zeroline=False, showticklabels=False), yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)))

fig.show()
```

```python
import plotly.graph_objs as go
import networkx as nx

# Assuming G is your NetworkX graph object
group_mapping = partidos.set_index('nome')['grupo'].to_dict()

# Create a graph from the similarity matrix dataframe

group_colors = {
    'oposicao': 'blue',
    'governo': 'red',
    'centrao_mdb': 'green',
    'centrao_uniao': 'yellow',
    'outros': 'grey'  # Default color for any group not listed above
}
# Assign the 'group' attribute from your 'partidos' dataframe to each node in the graph
for node in G.nodes():
    # If the node's name exists in the group mapping dictionary, assign the group value
    if node in group_mapping:
        G.nodes[node]['group'] = group_mapping[node]
    else:
        # If there's no group found for the node, assign a default value
        G.nodes[node]['group'] = 'outros'
    


# Generate edge traces
edge_trace = []
for edge in G.edges():
    char_1 = edge[0]
    char_2 = edge[1]
    x0, y0 = pos[char_1]
    x1, y1 = pos[char_2]
    trace = go.Scatter(x=[x0, x1, None], y=[y0, y1, None],
                       mode='lines',
                       line={'width': 0.5},
                       line_shape='spline',
                       opacity=0.5)
    edge_trace.append(trace)

# Generate node trace
node_trace = go.Scatter(x=[], y=[], 
                        text=[], 
                        mode='markers', 
                        textposition="bottom center",
                        hoverinfo='text', 
                        marker={'size':[(node_sizes[node]+1)*10 for node in G.nodes()], 'color':[]},
                        opacity=0.8)

for node in G.nodes():
    x, y = pos[node]
    node_trace['x'] += tuple([x])
    node_trace['y'] += tuple([y])
    node_trace['marker']['color'] += (group_colors[G.nodes[node]['group']],)  # Node color
    node_trace['text'] += tuple(['<b>' + node + '</b>'])
  # Node size

# Create figure
fig = go.Figure(data=edge_trace + [node_trace],
                layout=go.Layout(
                    showlegend=False, 
                    hovermode='closest',
                    margin=dict(b=0,l=0,r=0,t=0),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    clickmode='event+select',
                ))




fig.show()
```

```python
import dash
from dash import dcc, html, Input, Output, callback
import dash_cytoscape as cyto
import networkx as nx

nodes=[]

for node in G.nodes:
    nodes.append({
        'data': {'id': str(node), 'label': f'Node {node}', 'group': G.nodes[node]['group']},
        'position': {'x': pos[node][0] * 100, 'y': pos[node][1] * 100},
    })
# # Position nodes using NetworkX's layout
# pos = nx.spring_layout(G, k=0.1, iterations=50)
def get_elements(group_chosen, selected_node=None):
    elements = []

    # Filter nodes by group if specified
    nodes_chosen = G.nodes() if group_chosen == 'all' else [node for node in G.nodes() if G.nodes[node]['group'] == group_chosen]

    for node in nodes_chosen:
        elements.append({
            'data': {'id': str(node), 'label': f'Node {node}', 'group': G.nodes[node]['group']},
            'position': {'x': pos[node][0] * 100, 'y': pos[node][1] * 100},
        })

    # Include edges connected to the selected node or all edges if no node selected
    if selected_node:
        edges = [edge for edge in G.edges() if selected_node in edge]
    else:
        edges = G.edges()

    for edge in edges:
        if edge[0] in nodes_chosen and edge[1] in nodes_chosen:
            elements.append({'data': {'source': str(edge[0]), 'target': str(edge[1])}})

    return elements

app = dash.Dash(__name__)
elemnts = get_elements('all', None)
app.layout = html.Div([
    dcc.RadioItems(
        options=list(group_colors.keys()) + ['all'],
        value='all',
        id='controls-and-radio-item'
    ),
    cyto.Cytoscape(
        id='network-graph',
        elements=elemnts,
        style={'width': '100%', 'height': '500px'},
        layout={
            'name': 'preset',
            'positions': {
                node['data']['id']: nodes['position']
                for node in nodes
            }},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'label': 'data(label)',
                    'background-color': node['data']['group'],
                }
            },
            {
                'selector': 'edge',
                'style': {'width': 1, 'line-color': '#ccc'}
            },
        ]
    )
])

@app.callback(
    Output('network-graph', 'elements'),
    [Input('controls-and-radio-item', 'value'), Input('network-graph', 'tapNodeData')]
)
def update_graph(group_chosen, selected_node):
    selected_node_id = str(selected_node['id']) if selected_node else None
    return get_elements(group_chosen, selected_node_id)

if __name__ == '__main__':
    app.run_server(debug=True)
```
